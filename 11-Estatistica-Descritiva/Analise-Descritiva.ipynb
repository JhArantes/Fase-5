# Analise Descritiva

### Modulos


```python
import pandas as pd
import numpy as np
```

### Dados

```python
df_paises = pd.read_csv('../dados/Atividade cap 11 dados_1997_2011_paises_csv.csv', sep=';', decimal=',', encoding='latin1')

df_paises.head()
```

```python
df_paises.dtypes
```

- Contagem dos Paises

```python
df_paises.groupby('pais') \
    .agg(n = pd.NamedAgg('pais', 'count')) \
    .reset_index()
```

- Medidas descritivas: Posição

```python
df_paises.groupby('pais') \
    .agg(min_idh = pd.NamedAgg('idh', 'min'),
         max_idh = pd.NamedAgg('idh', 'max'),
         media_idh = pd.NamedAgg('idh', 'mean')) \
    .reset_index()
```

- Moda

```python
df_paises.groupby('pais')['idh'] \
    .apply(lambda x: x.mode().iloc[0]) \
    .to_frame() \
    .reset_index() 
```

- Mediana

```python
df_paises.groupby('pais') \
    .agg(median_idh = pd.NamedAgg('idh', 'median')) \
    .reset_index() 
```

### Percentis

```python
def my_quantile(array, q):
    n = len(array)
    index = (n - 1) * q
    if int(index) == index:
        return array[index]
    fraction = index - int(index)
    left = int(index)
    right = left + 1
    i, j = array[left], array[right]
    return i + (j - i) * fraction
```

```python
dados = [1, 1, 1, 2, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
my_quantile(dados, 0.35)
```

```python
np.quantile(dados, 0.35)
```

# Medidas de Variabilidade

- Desvio Médio Absoluto

```python
df_paises.groupby('pais') \
    .agg(desvio_medio_abs_idh = pd.NamedAgg('idh', 'std')) \
    .reset_index()
```

- Variancia e Desvio Padrão

```python
df_paises.groupby('pais') \
    .agg(variancia_idh = pd.NamedAgg('idh', 'var'),
         dp_idh = pd.NamedAgg('idh', 'std')) \
    .reset_index()
```
